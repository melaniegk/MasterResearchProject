```python
import os
import pandas as pd

# Function to convert sleep duration columns from minutes to hours
def convert_sleep_duration_to_hours(input_file, output_file):
    # Load data into a Pandas DataFrame
    df = pd.read_csv(input_file)

    # Convert sleep duration columns to hours (assuming column names)
    df['deepSleepTime (hours)'] = round(df['deepSleepTime'] / 60, 2)
    df['shallowSleepTime (hours)'] = round(df['shallowSleepTime'] / 60, 2)
    df['wakeTime (hours)'] = round(df['wakeTime'] / 60, 2)
    df['REMTime (hours)'] = round(df['REMTime'] / 60, 2)
    df['sleep_duration (hours)'] = round(df['sleep_duration'] / 60, 2)

    # Save the modified DataFrame to a new file
    df.to_csv(output_file, index=False)

# List of files in the folder
folder_path = 'C:/Users/User/Downloads/trial-1/B - MORE THAN 6'
files = os.listdir(folder_path)

# Loop through each file in the folder
for file in files:
    if file.endswith('.csv'):  # Assuming the files are CSV files
        input_file = os.path.join(folder_path, file)
        output_file = os.path.join(folder_path, 'converted_' + file)  # Output file name

        convert_sleep_duration_to_hours(input_file, output_file)
```
