```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro
from scipy.stats import normaltest

data = pd.read_csv('C:/Users/User/Downloads/sleep,hr,oxy participants list- demographic.xlsx - Sheet1.csv')

# Select columns 6 to 23
selected_columns = data.iloc[:, 6:24]

# Iterate over each column
for column in selected_columns.columns:
    # Extract the column data
    column_data = selected_columns[column].dropna()  # Remove any NaN values
    
    # Shapiro-Wilk Test for normality
    shapiro_test_stat, shapiro_p_value = shapiro(column_data)
    
    # D'Agostino and Pearson's Test for normality
    k2_stat, k2_p_value = normaltest(column_data)
    
    # Print results
    print(f"\nColumn: {column}")
    print(f"Shapiro-Wilk Test: Statistic={shapiro_test_stat}, p-value={shapiro_p_value}")
    print(f"D'Agostino and Pearson's Test: Statistic={k2_stat}, p-value={k2_p_value}")
    
    # Visualize the distribution
    plt.figure(figsize=(12, 6))
    
    # Histogram with KDE line
    plt.subplot(1, 2, 1)
    sns.histplot(column_data, kde=True, color='blue', alpha=0.7)
    plt.title('Histogram with KDE')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    
    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=column_data, color='green')
    plt.title('Boxplot')
    plt.xlabel(column)
    
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import pandas as pd
from scipy.stats import wilcoxon
from pingouin import compute_effsize

data = pd.read_csv('C:/Users/User/Downloads/sleep,hr,oxy participants list- demographic.xlsx - Sheet1.csv')

psg_deep_sleep = data['deepsleep_psg']
xiaomi_deep_sleep = data['deepsleep_xm']

# Calculate the observed median difference
observed_median_difference = np.median(psg_deep_sleep) - np.median(xiaomi_deep_sleep)

# Number of bootstrap samples
n_boot = 10000

# Initialize an array to store bootstrap median differences
bootstrap_differences = np.zeros(n_boot)

# Bootstrap resampling
for i in range(n_boot):
    # Sample with replacement from the differences
    resampled_differences = np.random.choice(psg_deep_sleep - xiaomi_deep_sleep, len(psg_deep_sleep), replace=True)
    # Calculate the median difference for the resampled data
    bootstrap_differences[i] = np.median(resampled_differences)

# Calculate the 95% confidence interval
ci_lower = np.percentile(bootstrap_differences, 2.5)
ci_upper = np.percentile(bootstrap_differences, 97.5)

# Perform Wilcoxon signed-rank test
stat, p_value = wilcoxon(psg_deep_sleep, xiaomi_deep_sleep)

# Calculate effect size (Vargha and Delaney's A)
eff_size = compute_effsize(psg_deep_sleep, xiaomi_deep_sleep, paired=True)

# Print results
print(f"Wilcoxon Signed-Rank Test for Deep Sleep:")
print(f"Test Statistic: {stat}, p-value: {p_value}")
print(f"Effect Size (Vargha and Delaney's A): {eff_size}")
print(f"Observed Median Difference (PSG - Xiaomi): {observed_median_difference}")
print(f"95% Confidence Interval for Median Difference: ({ci_lower}, {ci_upper})")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro
from scipy.stats import normaltest

data = pd.read_csv('C:/Users/User/Downloads/steps participants list- demographic.xlsx - Sheet1.csv')

# Select columns 6 to 23
selected_columns = data.iloc[:, 6:10]

# Iterate over each column
for column in selected_columns.columns:
    # Extract the column data
    column_data = selected_columns[column].dropna()  # Remove any NaN values
    
    # Shapiro-Wilk Test for normality
    shapiro_test_stat, shapiro_p_value = shapiro(column_data)
    
    # D'Agostino and Pearson's Test for normality
    k2_stat, k2_p_value = normaltest(column_data)
    
    # Print results
    print(f"\nColumn: {column}")
    print(f"Shapiro-Wilk Test: Statistic={shapiro_test_stat}, p-value={shapiro_p_value}")
    print(f"D'Agostino and Pearson's Test: Statistic={k2_stat}, p-value={k2_p_value}")
    
    # Visualize the distribution
    plt.figure(figsize=(12, 6))
    
    # Histogram with KDE line
    plt.subplot(1, 2, 1)
    sns.histplot(column_data, kde=True, color='blue', alpha=0.7)
    plt.title('Histogram with KDE')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    
    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=column_data, color='green')
    plt.title('Boxplot')
    plt.xlabel(column)
    
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import pandas as pd

data = pd.read_csv('C:/Users/User/Downloads/sleep,hr,oxy participants list- demographic.xlsx - Sheet1.csv')

psg_deep_sleep = data['deepsleep_psg']
xiaomi_deep_sleep = data['deepsleep_xm']

# Calculate median difference (bias)
median_difference = np.median(psg_deep_sleep - xiaomi_deep_sleep)

# Calculate interquartile range (IQR) as a measure of dispersion
iqr_of_differences = np.percentile(psg_deep_sleep - xiaomi_deep_sleep, 75) - np.percentile(psg_deep_sleep - xiaomi_deep_sleep, 25)

# Calculate 95% confidence interval for the median difference using the median absolute deviation (MAD)
mad = np.median(np.abs(psg_deep_sleep - xiaomi_deep_sleep - np.median(psg_deep_sleep - xiaomi_deep_sleep)))
ci_median_lower = median_difference - 1.96 * (1.57 * mad / np.sqrt(len(psg_deep_sleep)))
ci_median_upper = median_difference + 1.96 * (1.57 * mad / np.sqrt(len(psg_deep_sleep)))

# Calculate Bland-Altman 95% limits of agreement using the median and IQR
limits_of_agreement_lower = median_difference - 1.96 * (1.34 * iqr_of_differences)
limits_of_agreement_upper = median_difference + 1.96 * (1.34 * iqr_of_differences)

# Print results
print(f"Median Difference (Bias) for deep_sleep: {median_difference}")
print(f"Interquartile Range (IQR) of Differences for deep_sleep: {iqr_of_differences}")
print(f"95% Confidence Interval for Median Difference for deep_sleep: ({ci_median_lower}, {ci_median_upper})")
print(f"Bland-Altman 95% Limits of Agreement (using Median and IQR) for deep_sleep: ({limits_of_agreement_lower}, {limits_of_agreement_upper})")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the differences
differences = psg_deep_sleep - xiaomi_deep_sleep

# Bland-Altman Plot
plt.figure(figsize=(6, 4))
sns.scatterplot(x=np.median([psg_deep_sleep, xiaomi_deep_sleep], axis=0), y=differences)
plt.axhline(y=median_difference, color='r', linestyle='--', label='Median Difference (Bias)')
plt.axhline(y=limits_of_agreement_lower, color='g', linestyle='--', label='Lower Limit of Agreement')
plt.axhline(y=limits_of_agreement_upper, color='g', linestyle='--', label='Upper Limit of Agreement')
plt.xlabel('Median of Measurements')
plt.ylabel(f'Differences (PSG - Xiaomi) for Deep Sleep')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.title(f'Bland-Altman Plot for Deep Sleep')
plt.show()
```

```python
import numpy as np
import pandas as pd
import scipy.stats as stats

data = pd.read_csv('C:/Users/User/Downloads/sleep,hr,oxy participants list- demographic.xlsx - Sheet1.csv')

psg_shallowsleep = data['shallowsleep_PSG']
xiaomi_shallowsleep = data['shallowsleep_xm']

# Calculate mean difference (bias)
mean_difference = np.mean(psg_shallowsleep - xiaomi_shallowsleep)

# Calculate standard deviation of observed differences
sd_of_differences = np.std(psg_shallowsleep - xiaomi_shallowsleep, ddof=1)

# Calculate 95% confidence interval for the mean difference
t_value = stats.t.ppf(0.975, len(psg_shallowsleep) - 1)
ci_lower = mean_difference - t_value * (sd_of_differences / np.sqrt(len(psg_shallowsleep)))
ci_upper = mean_difference + t_value * (sd_of_differences / np.sqrt(len(psg_shallowsleep)))

# Calculate Bland-Altman 95% limits of agreement
limits_of_agreement_lower = mean_difference - 1.96 * sd_of_differences
limits_of_agreement_upper = mean_difference + 1.96 * sd_of_differences

# Print results
print(f"Mean Difference (Bias) for shallowsleep: {mean_difference}")
print(f"Standard Deviation of Observed Differences for shallowsleep: {sd_of_differences}")
print(f"95% Confidence Interval for shallowsleep: ({ci_lower}, {ci_upper})")
print(f"Bland-Altman 95% Limits of Agreement for shallowsleep: ({limits_of_agreement_lower}, {limits_of_agreement_upper})")
```

```python
import matplotlib.pyplot as plt

# Calculate the differences
differences = psg_shallowsleep - xiaomi_shallowsleep

# Bland-Altman Plot
plt.figure(figsize=(6, 4))
plt.scatter(x=np.mean([psg_shallowsleep, xiaomi_shallowsleep], axis=0), y=differences)
plt.axhline(y=mean_difference, color='r', linestyle='--', label='Mean Difference (Bias)')
plt.axhline(y=limits_of_agreement_lower, color='g', linestyle='--', label='Lower Limit of Agreement')
plt.axhline(y=limits_of_agreement_upper, color='g', linestyle='--', label='Upper Limit of Agreement')
plt.xlabel('Mean of Measurements')
plt.ylabel('Differences (PSG - Xiaomi) for Shallow Sleep')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.title('Bland-Altman Plot for Shallow Sleep')
plt.show()
```
